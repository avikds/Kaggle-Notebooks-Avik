```python
# =========================
# CONNECT X
# =========================

submission_code = r'''
import numpy as np
import random

def agent(observation, configuration):

    board = observation.board
    mark = observation.mark
    rows = configuration.rows
    cols = configuration.columns
    inarow = configuration.inarow
    opp = 1 if mark == 2 else 2

    def drop_piece(board, col, mark):
        b = board.copy()
        for r in range(rows - 1, -1, -1):
            if b[r * cols + col] == 0:
                b[r * cols + col] = mark
                break
        return b

    def valid_moves(board):
        return [c for c in range(cols) if board[c] == 0]

    def check_winner(board, m):
        grid = np.array(board).reshape(rows, cols)

        for r in range(rows):
            for c in range(cols - inarow + 1):
                if np.all(grid[r, c:c+inarow] == m):
                    return True

        for c in range(cols):
            for r in range(rows - inarow + 1):
                if np.all(grid[r:r+inarow, c] == m):
                    return True

        for r in range(rows - inarow + 1):
            for c in range(cols - inarow + 1):
                if all(grid[r+i, c+i] == m for i in range(inarow)):
                    return True

        for r in range(inarow - 1, rows):
            for c in range(cols - inarow + 1):
                if all(grid[r-i, c+i] == m for i in range(inarow)):
                    return True

        return False

    def minimax(board, depth, alpha, beta, maximizing):
        moves = valid_moves(board)

        if depth == 0 or check_winner(board, mark) or check_winner(board, opp) or not moves:
            if check_winner(board, mark):
                return None, 1e6
            if check_winner(board, opp):
                return None, -1e6
            return None, 0

        if maximizing:
            best = -1e9
            col = random.choice(moves)
            for c in moves:
                val = minimax(drop_piece(board, c, mark), depth-1, alpha, beta, False)[1]
                if val > best:
                    best = val
                    col = c
                alpha = max(alpha, best)
                if alpha >= beta:
                    break
            return col, best
        else:
            best = 1e9
            col = random.choice(moves)
            for c in moves:
                val = minimax(drop_piece(board, c, opp), depth-1, alpha, beta, True)[1]
                if val < best:
                    best = val
                    col = c
                beta = min(beta, best)
                if alpha >= beta:
                    break
            return col, best

    if board.count(0) == rows * cols:
        return cols // 2

    return minimax(board, 3, -1e9, 1e9, True)[0]
'''

# Write submission.py
with open("submission.py", "w") as f:
    f.write(submission_code)

print("submission.py written successfully")
```
