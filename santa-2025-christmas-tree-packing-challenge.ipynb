```python
# ============================================================
# Santa 2025 - Christmas Tree Packing Challenge
# ============================================================

import math
import pandas as pd
from shapely import affinity
from shapely.geometry import Polygon
from shapely.ops import unary_union
from shapely.strtree import STRtree

# ------------------------------------------------------------
# Christmas tree polygon
# ------------------------------------------------------------

def make_tree(x, y, deg):
    trunk_w = 0.15
    trunk_h = 0.2
    base_w = 0.7
    mid_w = 0.4
    top_w = 0.25
    tip_y = 0.8
    tier_1_y = 0.5
    tier_2_y = 0.25
    base_y = 0.0
    trunk_bottom_y = -trunk_h

    pts = [
        (0.0, tip_y),
        (top_w/2, tier_1_y), (top_w/4, tier_1_y),
        (mid_w/2, tier_2_y), (mid_w/4, tier_2_y),
        (base_w/2, base_y),
        (trunk_w/2, base_y), (trunk_w/2, trunk_bottom_y),
        (-trunk_w/2, trunk_bottom_y), (-trunk_w/2, base_y),
        (-base_w/2, base_y),
        (-mid_w/4, tier_2_y), (-mid_w/2, tier_2_y),
        (-top_w/4, tier_1_y), (-top_w/2, tier_1_y),
    ]

    poly = Polygon(pts)
    poly = affinity.rotate(poly, deg, origin=(0, 0))
    poly = affinity.translate(poly, xoff=x, yoff=y)
    return poly


# ------------------------------------------------------------
# Collision check
# ------------------------------------------------------------

def no_overlap(polys):
    tree = STRtree(polys)
    for i, p in enumerate(polys):
        for j in tree.query(p):
            if i != j and p.intersects(polys[j]) and not p.touches(polys[j]):
                return False
    return True


# ------------------------------------------------------------
# Optimize layout for given n
# ------------------------------------------------------------

def optimize_n(n):
    best_side = float("inf")
    best_layout = None

    rotations = [0.0, 90.0, 180.0, 270.0]

    spacing = 1.25
    while spacing > 0.75:
        grid = math.ceil(math.sqrt(n))
        coords = []

        for i in range(grid):
            for j in range(grid):
                if len(coords) < n:
                    coords.append((i * spacing, j * spacing))

        for rot in rotations:
            polys = [make_tree(x, y, rot) for x, y in coords]

            if not no_overlap(polys):
                continue

            bounds = unary_union(polys).bounds
            side = max(bounds[2] - bounds[0], bounds[3] - bounds[1])

            if side < best_side:
                best_side = side
                best_layout = [(x, y, rot) for x, y in coords]

        spacing -= 0.03

    return best_layout


# ------------------------------------------------------------
# Run full optimization
# ------------------------------------------------------------

rows = []

for n in range(1, 201):
    print(f"Optimizing n={n}")
    layout = optimize_n(n)

    xs = [x for x, _, _ in layout]
    ys = [y for _, y, _ in layout]

    cx = (min(xs) + max(xs)) / 2
    cy = (min(ys) + max(ys)) / 2

    for i, (x, y, deg) in enumerate(layout):
        rows.append({
            "id": f"{n:03d}_{i}",
            "x": f"s{(x - cx):.8f}",
            "y": f"s{(y - cy):.8f}",
            "deg": f"s{deg:.6f}",
        })

submission = pd.DataFrame(rows, columns=["id", "x", "y", "deg"])
submission.to_csv("submission.csv", index=False)

print("submission.csv created")
```
