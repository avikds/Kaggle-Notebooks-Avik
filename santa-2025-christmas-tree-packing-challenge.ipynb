```python
# ============================================================
# Santa 2025 - Christmas Tree Packing Challenge
# ============================================================

import math
import pandas as pd

# -------------------------
# Configuration
# -------------------------
MAX_TREES = 200

# Safe but dense spacing (no rotation)
DX = 0.92
DY = DX * math.sqrt(3) / 2  # hex vertical spacing

ROTATION = 0.0

# -------------------------
# Generate layouts
# -------------------------
rows = []

for n in range(1, MAX_TREES + 1):
    # estimate rows needed
    rows_needed = math.ceil(math.sqrt(n))
    
    coords = []
    idx = 0

    for r in range(rows_needed * 2):
        y = r * DY
        offset = (DX / 2) if (r % 2 == 1) else 0.0

        for c in range(rows_needed * 2):
            x = c * DX + offset
            coords.append((x, y))
            idx += 1
            if idx >= n:
                break
        if idx >= n:
            break

    # center the configuration
    xs = [p[0] for p in coords[:n]]
    ys = [p[1] for p in coords[:n]]

    cx = (min(xs) + max(xs)) / 2
    cy = (min(ys) + max(ys)) / 2

    for i, (x, y) in enumerate(coords[:n]):
        rows.append({
            "id": f"{n:03d}_{i}",
            "x": f"s{(x - cx):.8f}",
            "y": f"s{(y - cy):.8f}",
            "deg": f"s{ROTATION:.6f}",
        })

# -------------------------
# Save submission
# -------------------------
submission = pd.DataFrame(rows, columns=["id", "x", "y", "deg"])
submission.to_csv("submission.csv", index=False)

print("✅ submission.csv created")
print("Rows:", len(submission))
submission.head()
```
