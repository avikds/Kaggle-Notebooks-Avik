```python
# ============================================================
# Santa 2025 - Christmas Tree Packing Challenge
# Fast baseline solution (grid packing)
# ============================================================

import math
import pandas as pd

# -------------------------
# Configuration
# -------------------------
MAX_TREES = 200
TREE_SPACING = 0.85   # safe spacing to avoid collisions
ROTATIONS = [0.0, 180.0]  # alternate rotations

# -------------------------
# Generate packing
# -------------------------
rows = []

for n in range(1, MAX_TREES + 1):
    # grid size (square)
    grid_size = math.ceil(math.sqrt(n))

    # total side length
    side = (grid_size - 1) * TREE_SPACING

    # start so that center is near (0,0)
    x0 = -side / 2
    y0 = -side / 2

    idx = 0
    for i in range(grid_size):
        for j in range(grid_size):
            if idx >= n:
                break

            x = x0 + i * TREE_SPACING
            y = y0 + j * TREE_SPACING
            deg = ROTATIONS[idx % 2]

            rows.append({
                "id": f"{n:03d}_{idx}",
                "x": f"s{x:.6f}",
                "y": f"s{y:.6f}",
                "deg": f"s{deg:.6f}"
            })

            idx += 1
        if idx >= n:
            break

# -------------------------
# Create submission file
# -------------------------
submission = pd.DataFrame(rows, columns=["id", "x", "y", "deg"])
submission.to_csv("submission.csv", index=False)

print("✅ Submission file created: submission.csv")
print("Total rows:", len(submission))
submission.head()
```
