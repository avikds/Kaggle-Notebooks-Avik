```python
# ============================================================
# Santa 2025 - Christmas Tree Packing Challenge
# ============================================================

import math
import pandas as pd

MAX_TREES = 200

# Safe spacing (tight but collision-free without rotation)
SPACING = 0.95
ROT = 0.0

# Generate square spiral coordinates
def square_spiral(n, spacing):
    coords = [(0.0, 0.0)]
    if n == 1:
        return coords

    step = 1
    x, y = 0, 0
    while len(coords) < n:
        # right
        for _ in range(step):
            x += spacing
            coords.append((x, y))
            if len(coords) == n:
                return coords
        # up
        for _ in range(step):
            y += spacing
            coords.append((x, y))
            if len(coords) == n:
                return coords
        step += 1
        # left
        for _ in range(step):
            x -= spacing
            coords.append((x, y))
            if len(coords) == n:
                return coords
        # down
        for _ in range(step):
            y -= spacing
            coords.append((x, y))
            if len(coords) == n:
                return coords
        step += 1

    return coords


rows = []

for n in range(1, MAX_TREES + 1):
    coords = square_spiral(n, SPACING)

    xs = [p[0] for p in coords]
    ys = [p[1] for p in coords]

    cx = (min(xs) + max(xs)) / 2
    cy = (min(ys) + max(ys)) / 2

    for i, (x, y) in enumerate(coords):
        rows.append({
            "id": f"{n:03d}_{i}",
            "x": f"s{(x - cx):.8f}",
            "y": f"s{(y - cy):.8f}",
            "deg": f"s{ROT:.6f}",
        })

submission = pd.DataFrame(rows, columns=["id", "x", "y", "deg"])
submission.to_csv("submission.csv", index=False)

print("✅ submission.csv created")
print("Rows:", len(submission))
submission.head()
```
