```python
# ============================================================
# Santa 2025 - Christmas Tree Packing Challenge
# ============================================================

import math
import pandas as pd

# -------------------------
# Configuration
# -------------------------
MAX_TREES = 200

# SAFE spacing (empirically + geometrically safe)
TREE_SPACING = 1.05

# No rotation = predictable bounding
ROTATION_DEG = 0.0

# -------------------------
# Generate placements
# -------------------------
rows = []

for n in range(1, MAX_TREES + 1):
    grid = math.ceil(math.sqrt(n))
    side = (grid - 1) * TREE_SPACING

    x_start = -side / 2
    y_start = -side / 2

    idx = 0
    for i in range(grid):
        for j in range(grid):
            if idx >= n:
                break

            x = x_start + i * TREE_SPACING
            y = y_start + j * TREE_SPACING

            rows.append({
                "id": f"{n:03d}_{idx}",
                "x": f"s{x:.8f}",
                "y": f"s{y:.8f}",
                "deg": f"s{ROTATION_DEG:.6f}",
            })

            idx += 1
        if idx >= n:
            break

# -------------------------
# Save submission
# -------------------------
submission = pd.DataFrame(rows, columns=["id", "x", "y", "deg"])
submission.to_csv("submission.csv", index=False)

print("✅ submission.csv created successfully")
print("Rows:", len(submission))
submission.head()
```
